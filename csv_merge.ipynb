```python
import pandas as pd
import glob
import os

# 1. Define your folder path (using 'r' before the string to handle Windows backslashes)
folder_path = r"C:\Users\ASUS\Documents\Semester6\Big Data\Lazada Scrape\output"

# 2. Define the categories based on your file naming convention
categories = ["baju", "celana", "dress", "kemeja"]

# 3. Loop through each category to merge its files
for category in categories:
    print(f"🚀 Processing category: {category}...")
    
    # Create a search pattern (e.g., lazada_baju_category_v*.csv)
    search_pattern = os.path.join(folder_path, f"lazada_{category}_category_v*.csv")
    
    # Find all files that match the pattern
    file_list = glob.glob(search_pattern)
    
    if not file_list:
        print(f"⚠️ No files found for {category}. Skipping...\n")
        continue
        
    print(f"Found {len(file_list)} files for {category}.")
    
    # Read each CSV file into a dataframe and store them in a list
    dataframes = []
    for file in file_list:
        try:
            df = pd.read_csv(file)
            dataframes.append(df)
        except Exception as e:
            print(f"❌ Error reading {file}: {e}")
            
    # Combine all dataframes in the list into one master dataframe
    if dataframes:
        merged_df = pd.concat(dataframes, ignore_index=True)
        
        # Define the output file name
        output_file = os.path.join(folder_path, f"lazada_{category}.csv")
        
        # Save the merged dataframe to CSV
        merged_df.to_csv(output_file, index=False)
        print(f"✅ Successfully saved: {output_file}\n")
    else:
        print(f"⚠️ No valid data to merge for {category}.\n")

print("🎉 All merging tasks completed!")
```

```python
#merge all csv files in the output folder into one master file
import pandas as pd
import os

# 1. Define your folder path
folder_path = r"C:\Users\ASUS\Documents\Semester6\Big Data\Lazada Scrape\output"

# 2. List the four category files you want to merge
files_to_merge = [
    "lazada_kemeja.csv",
    "lazada_baju.csv",
    "lazada_dress.csv",
    "lazada_celana.csv"
]

# 3. Read each file and store it in a list
dataframes = []
for file in files_to_merge:
    file_path = os.path.join(folder_path, file)
    try:
        df = pd.read_csv(file_path)
        dataframes.append(df)
        print(f"✅ Loaded: {file} (Rows: {len(df)})")
    except FileNotFoundError:
        print(f"❌ Error: Could not find {file}. Check if the previous script saved it correctly.")
    except Exception as e:
        print(f"❌ Unexpected error reading {file}: {e}")

# 4. Combine them all and save the final file
if dataframes:
    print("\n⏳ Merging all categories together...")
    final_df = pd.concat(dataframes, ignore_index=True)
    
    # Define the final output file name
    output_file = os.path.join(folder_path, "lazada_reviews.csv")
    
    # Save to CSV
    final_df.to_csv(output_file, index=False)
    
    print(f"🎉 Success! Final dataset saved as: {output_file}")
    print(f"📊 Total rows in lazada_reviews.csv: {len(final_df)}")
else:
    print("\n⚠️ No data was loaded. Please check your files.")
```
